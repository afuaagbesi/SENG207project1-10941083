###  Replacement method

```python
Columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol', 'quality']
```

```python
for column in Columns:
    data[column].fillna(data[column].mean(), inplace=True)
```

 When dealing with missing values in data analysis, it is common to use the mean of the column to replace null values instead of removing rows with missing values. This is because removing rows can lead to a loss of information and a reduction in the sample size, which can negatively affect the accuracy and representativeness of the analysis. To avoid this, replacing null values with the mean of the column can retain the sample size and provide a reasonable estimate of the missing data. However, this method assumes that the missing values are randomly distributed and that the data follows a normal distribution. If these assumptions are not met, using the mean may not accurately represent the missing data. In such cases, other techniques like interpolation, imputation, or removing rows/columns may be more appropriate.

```python
mean_quality=data['quality'].mean()
compared_mean=data['quality']<mean_quality
mpl.pie(compared_mean.value_counts(),labels=['Values less than mean','Values greater than mean'])
mpl.title('QUality values compared to the mean') 
mpl.show()   
```
